```python
import numpy as np
```

```python
i1 = 0.05
i2 = 0.10
inputs = np.array([i1, i2])
```

```python
W_hidden = np.random.uniform(-0.5, 0.5, (2, 2)) #input to hidden
W_output = np.random.uniform(-0.5, 0.5, (2, 2)) #hidden to output
```

```python
b1, b2= 0.5, 0.7
```

```python
def tanh(x):
    return np.tanh(x)
    #Hidden_Layers
hidden_values= np.dot(W_hidden,inputs)+b1
h=tanh(hidden_values)
    #Output_Layer
output_values = np.dot(W_output,h)+b2
o=tanh(output_values)
```

```python
print("Hidden Layer Output (h1, h2):", h)
print("Final Output (o1, o2):", o)
```
